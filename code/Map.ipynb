```python
import geopandas as gp
import matplotlib.pyplot as plt
```

```python
states = gp.read_file("../shape/states.shp")
```

```python
gdf = gp.GeoDataFrame([], geometry=gp.points_from_xy(y=[37.384464], x=[-80.097837]))


f, ax = plt.subplots(1, figsize = (8,5))
states[states['SUB_REGION'].isin(['New England', 'West North Central',
       'East North Central', 'Middle Atlantic', 'South Atlantic',
       'East South Central', 'West South Central'])].plot(facecolor='none', edgecolor='black', ax = ax)
states[states['STATE_NAME'] == "Virginia"].plot(facecolor='black',edgecolor = 'none',linewidth = 5, ax = ax)
gdf.plot(ax = ax, marker = "*", markersize = 220, facecolor='red')
ax.tick_params(axis='x', labelsize=20)
ax.tick_params(axis='y', labelsize=20)
plt.grid()
plt.title("Study Area: Blacksburg, VA, USA", fontsize = 30)
```
