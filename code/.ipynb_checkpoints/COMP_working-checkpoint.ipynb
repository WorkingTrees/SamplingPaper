```python
# Histograms of error by species
# Convert to carbon
# F test and other error metrics 
```

```python
# Jenkins equtions For mixed hardwoods 
def biom_f_dbh(dbh, treetype = 'mixed hardwoods'):
    
    dbh_cm = dbh* 2.54
    
    if treetype == 'mixed hardwoods':
        return np.exp(-2.4800  + 2.4835 *np.log(dbh_cm))
    if treetype == 'pine':
        return np.exp(-2.5356  + 2.4349 *np.log(dbh_cm))
    else:
        return "Specify species"
```

```python
# Apply jenkins equations for each of the 3 species

bw = compdf[compdf['species'] == 'Black Walnut']
bw['biom_app'] = biom_f_dbh(bw['diam_app'], treetype = 'mixed hardwoods')
bw['biom'] = biom_f_dbh(bw['diam'], treetype = 'mixed hardwoods')

bl = compdf[compdf['species'] == 'Black Locust']
bl['biom_app'] = biom_f_dbh(bl['diam_app'], treetype = 'mixed hardwoods')
bl['biom'] = biom_f_dbh(bl['diam'], treetype = 'mixed hardwoods')

llp = compdf[compdf['species'] == 'Loblolly Pine']
llp['biom_app'] = biom_f_dbh(llp['diam_app'], treetype = 'pine')
llp['biom'] = biom_f_dbh(llp['diam'], treetype = 'pine')
```

```python
# Concat the dfs
biomdf = pd.concat([bw,bl,llp], axis = 0)

# convert from woody biomass (kg) to tons
biomdf['biom_app'] = biomdf['biom_app'] * 0.001
biomdf['biom'] = biomdf['biom'] * 0.001
# convert from tons of woody biomass to CO2e 
biomdf['co2e_app'] = biomdf['biom_app'] * 0.45 * 44/12
biomdf['co2e'] = biomdf['biom'] * 0.45 * 44/12
```

```python
groups = biomdf.groupby('species')

# Plot
fig, ax = plt.subplots(figsize = (6,4))

for name, group in groups:
    ax.plot(group['co2e'], group['co2e_app'], marker='o', linestyle='', ms=8, label=name)
    
# Plot 1-1 line
ax.plot([0,300],[0,300],'k-', lw=2, label = '1-1 line')

# Compute regression 
slope, intercept, r_value, p_value, std_err = stats.linregress(biomdf['co2e'], biomdf['co2e_app'])

# regression line 
line = slope*np.linspace(0,300)+ intercept
plt.plot(np.linspace(0,300), line, 'r', label='y={:.2f}x+{:.2f}'.format(slope,intercept))
plt.annotate("$R^2$ = {}".format(str(r_value**2)[:5]), [0.25,0.02])

ax.legend()
ax.set_xlim(0,2.5)
ax.set_ylim(0,2.5)
    
plt.xlabel("co2e (Metric Tons)", size = 14)
plt.ylabel("app-based co2e (Metric Tons)", size = 14)

plt.title("co2e comparison", size = 20)
```

```python
for species in compdf['species'].unique():
    spdf = compdf[compdf['species'] == species]
    print(species, "n = {}".format(len(spdf)))
    print("app_height MEAN = {}".format(str(round(spdf['height_app'].mean(),3))))
    print("manual_height MEAN = {}".format(str(round(spdf['height_in'].mean(),3))))
    
    print("app_height VAR = {}".format(str(round(spdf['height_app'].var(),3))))
    print("manual_height VAR = {}".format(str(round(spdf['height_in'].var(),3))))
    
    print("app_height STDERR = {}".format(str(round(spdf['height_app'].std() / np.sqrt(len(spdf)),3))))
    print("manual_height STDERR = {}".format(str(round(spdf['height_in'].std() / np.sqrt(len(spdf)),3))))
    
    print("app_diam MEAN = {}".format(str(round(spdf['diam_app'].mean(),3))))
    print("manual_diam MEAN = {}".format(str(round(spdf['diam'].mean(),3))))
    
    print("app_diam VAR = {}".format(str(round(spdf['diam_app'].var(),3))))
    print("manual_diam VAR = {}".format(str(round(spdf['diam'].var(),3))))
    
    print("app_diam STDERR = {}".format(str(round(spdf['diam_app'].std() / np.sqrt(len(spdf)),3))))
    print("manual_diam STDERR = {}".format(str(round(spdf['diam'].std() / np.sqrt(len(spdf)),3))))
    
    print("T TEST DIAM")
    print(stats.ttest_ind(spdf['diam'], spdf['diam_app']))
    
    print("T TEST HEIGHT")
    print(stats.ttest_ind(spdf['height_in'], spdf['height_app']))
    
    print("SIGNED RANK TEST DIAM")
    print(stats.wilcoxon(spdf['diam'], spdf['diam_app']))
    
    print("SIGNED RANK TEST HEIGHT")
    print(stats.wilcoxon(spdf['height_in'], spdf['height_app']))
    
    
    print("====" * 15)
```
