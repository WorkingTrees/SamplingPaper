```python
def correct_outliers(df, threshold = 1.1):

    # Sort the dataframe by date
    df = df.sort_values(by=['date'])
    df['date'] = pd.to_datetime(session_df['date'])

    # Identify outliers based on latitude and longitude
    lat_mean = df['lat'].mean()
    lat_std = df['lat'].std()
    lon_mean = df['lon'].mean()
    lon_std = df['lon'].std()
    is_outlier = ((df['lat'] < lat_mean - threshold * lat_std) | (df['lat'] > lat_mean + threshold * lat_std) |
                  (df['lon'] < lon_mean - threshold * lon_std) | (df['lon'] > lon_mean + threshold * lon_std))

    df['outlier'] = is_outlier

    cleaned_rows =[]

    # Loop through, correct based on +/2 points in time 
    for idx, row in df.reset_index().iterrows():
        if row['outlier']:

            # check if in the last row, index for before if so. 
            if idx == len(df):
                df['timedelta'] = abs(df['date'] - row['date'])
                tmpdf = df[df['outlier'] == False].sort_values(by=['timedelta'], ascending = True)[0:5]
                lat_cor = np.median([tmpdf['lat']])
                lon_cor = np.median([tmpdf['lon']])

            # same if last-1 row
            elif idx == len(df) - 1:
                df['timedelta'] = abs(df['date'] - row['date'])
                tmpdf = df[df['outlier'] == False].sort_values(by=['timedelta'], ascending = True)[0:5]
                lat_cor = np.median([tmpdf['lat']])
                lon_cor = np.median([tmpdf['lon']])

            # if zeroeth row, index after
            elif idx == 0:
                df['timedelta'] = abs(df['date'] - row['date'])
                tmpdf = df[df['outlier'] == False].sort_values(by=['timedelta'], ascending = True)[0:5]
                lat_cor = np.median([tmpdf['lat']])
                lon_cor = np.median([tmpdf['lon']])

            # if not, index before/after
            else:
                df['timedelta'] = abs(df['date'] - row['date'])
                tmpdf = df[df['outlier'] == False].sort_values(by=['timedelta'], ascending = True)[0:5]
                lat_cor = np.median([tmpdf['lat']])
                lon_cor = np.median([tmpdf['lon']])

            # Apply the corrections 
            filtered = row.copy()
            filtered['lat'] = lat_cor
            filtered['lon'] = lon_cor
            cleaned_rows.append(pd.DataFrame(filtered).T)

        # If not outlier, keep the original row 
        else:
            cleaned_rows.append(pd.DataFrame(row).T)


    return pd.concat(cleaned_rows, axis = 0)
```

```python
# Apply the GPS filtering to remove spatial outliers 

clean_df = correct_outliers(session_df, threshold=1.1)

pldf = clean_df.copy()

geometry = [Point(xy) for xy in zip(pldf['lon'], pldf['lat'])]
plot_df = pldf.drop(['lat', 'lon'], axis=1)
gdf = gp.GeoDataFrame(plot_df, crs="EPSG:4326", geometry=geometry)

f, ax = plt.subplots()
gdf.plot(ax=ax, marker='o', column='parcel_id', categorical=True, legend=True)
leg = ax.get_legend()
leg.set_bbox_to_anchor((0., 0., 1.2, 1.2))

plt.show()
```

# Regular grid

```python

def create_polygon_grid(points_gdf, num_points_per_polygon=1, grid_scale = 1):
    """
    Creates a grid of polygons that covers all points in a geopandas geodataframe and contains
    approximately num_points_per_polygon points per polygon.
    
    Args:
    - points_gdf: geopandas geodataframe of points
    - num_points_per_polygon: min number of points to be contained in each polygon
    
    Returns:
    - geopandas geodataframe of polygons
    """
    
    # get the bounds of the points and calculate the grid size
    xmin, ymin, xmax, ymax = points_gdf.total_bounds
    x_range = xmax - xmin
    y_range = ymax - ymin
    
    # calculate the number of rows and columns in the grid
    num_rows = int(y_range / x_range * np.sqrt(len(points_gdf)) + 1)
    num_cols = int(x_range / y_range * np.sqrt(len(points_gdf)) + 1)
    
    # calculate the width and height of each cell in the grid
    cell_width = x_range / num_cols * grid_scale
    cell_height = y_range / num_rows * grid_scale
    
    # create an empty list to store the polygons
    polygons = []
    
    # iterate over each cell in the grid
    for i in range(num_rows):
        for j in range(num_cols):
            # calculate the bounds of the cell
            cell_xmin = xmin + j * cell_width
            cell_ymin = ymin + i * cell_height
            cell_xmax = cell_xmin + cell_width
            cell_ymax = cell_ymin + cell_height
            
            # create a polygon from the cell bounds
            cell_polygon = Polygon([(cell_xmin, cell_ymin), (cell_xmax, cell_ymin),
                                    (cell_xmax, cell_ymax), (cell_xmin, cell_ymax)])
            
            # count the number of points in the cell
            num_points_in_cell = len(points_gdf[points_gdf.within(cell_polygon)])
            
            # if the cell contains at least num_points_per_polygon points, add it to the list of polygons
            if num_points_in_cell >= num_points_per_polygon:
                polygons.append(cell_polygon)
                
    # create a geopandas geodataframe from the polygons
    polygons_gdf = gp.GeoDataFrame(geometry=polygons)
    
    return polygons_gdf
```

# Hex grid

```python
def create_hexgrid(points_gdf, avg_points_per_hex):
    # Calculate bounding box of points
    minx, miny, maxx, maxy = points_gdf.buffer(0.0001).total_bounds
    
    # Calculate hexagon side length based on average points per hexagon
    hex_area = ((maxx - minx) * (maxy - miny)) / len(points_gdf) * avg_points_per_hex
    side_length = np.sqrt(2 * hex_area / (3 * np.sqrt(3)))
    
    # Calculate hexagon grid using bounding box and side length
    x_step = 3 * side_length
    y_step = 1.5 * side_length
    x_span = maxx - minx
    y_span = maxy - miny
    x_cols = int(np.ceil(x_span / x_step))
    y_cols = int(np.ceil(y_span / y_step))
    hexagons = []
    hex_counts = []
    for x in range(x_cols):
        for y in range(y_cols):
            hex_x = x * x_step + minx + ((y % 2) * 1.5 * side_length)
            hex_y = y * y_step + miny
            hex_coords = [
                [hex_x, hex_y + side_length],
                [hex_x + 1.5 * side_length, hex_y + 0.5 * side_length],
                [hex_x + 1.5 * side_length, hex_y - 0.5 * side_length],
                [hex_x, hex_y - side_length],
                [hex_x - 1.5 * side_length, hex_y - 0.5 * side_length],
                [hex_x - 1.5 * side_length, hex_y + 0.5 * side_length],
                [hex_x, hex_y + side_length]
            ]
            hex_polygon = Polygon(hex_coords)
            hexagons.append(hex_polygon)
            hex_counts.append(points_gdf[points_gdf.geometry.within(hex_polygon)].shape[0])
    
    # Convert hexagons to GeoDataFrame and perform spatial join with points
    hex_gdf = gp.GeoDataFrame({'geometry': hexagons, 'points_count': hex_counts})
    hex_gdf.crs = points_gdf.crs
    hex_with_points = gp.sjoin(hex_gdf, points_gdf, op='contains')
    
    # Return only unique geometries
    unique_geoms = hex_with_points['geometry'].unique()
    unique_gdf = gp.GeoDataFrame(geometry=unique_geoms)
    unique_gdf = unique_gdf.merge(hex_gdf[['geometry', 'points_count']], on='geometry')
    unique_gdf.crs = hex_gdf.crs
    
    return unique_gdf
```

```python
# Jenkins equtions For mixed hardwoods 

def biom_f_dbh(dbh, treetype = 'mixed hardwoods'):
    
    dbh_cm = dbh* 2.54
    
    if treetype == 'mixed hardwoods':
        return np.exp(-2.4800  + 2.4835 *np.log(dbh_cm))
    if treetype == 'pine':
        return np.exp(-2.5356  + 2.4349 *np.log(dbh_cm))
    else:
        return "Specify species"
```

```python
def correct_outliers(df, threshold = 1.7):

    # Sort the dataframe by date
    df = df.sort_values(by=['date'])

    # Identify outliers based on latitude and longitude
    lat_mean = df['lat'].mean()
    lat_std = df['lat'].std()
    lon_mean = df['lon'].mean()
    lon_std = df['lon'].std()
    is_outlier = ((df['lat'] < lat_mean - threshold * lat_std) | (df['lat'] > lat_mean + threshold * lat_std) |
                  (df['lon'] < lon_mean - threshold * lon_std) | (df['lon'] > lon_mean + threshold * lon_std))

    df['outlier'] = is_outlier

    cleaned_rows =[]

    # Loop through, correct based on +/1 point in time 
    for idx, row in df.reset_index().iterrows():
        if row['outlier']:

            # check if in the last row, index for before if so. 
            if idx == len(df):
                df['timedelta'] = df['date'] - row['date']
                tmpdf = df[df['outlier'] == False].sort_values(by=['timedelta'], ascending = False)[0:3]
                lat_cor = np.mean([tmpdf.iloc[0]['lat'], tmpdf.iloc[1]['lat']])
                lon_cor = np.mean([tmpdf.iloc[0]['lon'], tmpdf.iloc[1]['lon']])

            # same if last-1 row
            elif idx == len(df) - 1:
                df['timedelta'] = df['date'] - row['date']
                tmpdf = df[df['outlier'] == False].sort_values(by=['timedelta'], ascending = False)[0:3]
                lat_cor = np.mean([tmpdf.iloc[0]['lat'], tmpdf.iloc[1]['lat']])
                lon_cor = np.mean([tmpdf.iloc[0]['lon'], tmpdf.iloc[1]['lon']])

            # if zeroeth row, index after
            elif idx == 0:
                lat_cor = np.mean([df.iloc[idx+2]['lat'], df.iloc[idx+1]['lat']])
                lon_cor = np.mean([df.iloc[idx+2]['lon'], df.iloc[idx+1]['lon']])

            # if not, index before/after
            else:
                lat_cor = np.mean([df.iloc[idx-1]['lat'], df.iloc[idx+1]['lat']])
                lon_cor = np.mean([df.iloc[idx-1]['lon'], df.iloc[idx+1]['lon']])

            # Apply the corrections 
            filtered = row.copy()
            filtered['lat'] = lat_cor
            filtered['lon'] = lon_cor
            cleaned_rows.append(pd.DataFrame(filtered).T)

        # If not outlier, keep the original row 
        else:
            cleaned_rows.append(pd.DataFrame(row).T)


    return pd.concat(cleaned_rows, axis = 0)

```

```python
def create_hexgrid(points_gdf, avg_points_per_hex):
    # Calculate bounding box of points
    minx, miny, maxx, maxy = points_gdf.buffer(0.0001).total_bounds
    
    # Calculate hexagon side length based on average points per hexagon
    hex_area = ((maxx - minx) * (maxy - miny)) / len(points_gdf) * avg_points_per_hex
    side_length = np.sqrt(2 * hex_area / (3 * np.sqrt(3)))
    
    # Calculate hexagon grid using bounding box and side length
    x_step = 3 * side_length
    y_step = 1.5 * side_length
    x_span = maxx - minx
    y_span = maxy - miny
    x_cols = int(np.ceil(x_span / x_step))
    y_cols = int(np.ceil(y_span / y_step))
    hexagons = []
    hex_counts = []
    for x in range(x_cols):
        for y in range(y_cols):
            hex_x = x * x_step + minx + ((y % 2) * 1.5 * side_length)
            hex_y = y * y_step + miny
            hex_coords = [
                [hex_x, hex_y + side_length],
                [hex_x + 1.5 * side_length, hex_y + 0.5 * side_length],
                [hex_x + 1.5 * side_length, hex_y - 0.5 * side_length],
                [hex_x, hex_y - side_length],
                [hex_x - 1.5 * side_length, hex_y - 0.5 * side_length],
                [hex_x - 1.5 * side_length, hex_y + 0.5 * side_length],
                [hex_x, hex_y + side_length]
            ]
            hex_polygon = Polygon(hex_coords)
            hexagons.append(hex_polygon)
            hex_counts.append(points_gdf[points_gdf.geometry.within(hex_polygon)].shape[0])
    
    # Convert hexagons to GeoDataFrame and perform spatial join with points
    hex_gdf = gp.GeoDataFrame({'geometry': hexagons, 'points_count': hex_counts})
    hex_gdf.crs = points_gdf.crs
    hex_with_points = gp.sjoin(hex_gdf, points_gdf, op='contains')
    
    # Create a dictionary of points within each hexagon
    points_by_hex = {}
    for i, row in hex_with_points.iterrows():
        hex_geom = row['geometry']
        point_idx = row.index_right
        if hex_geom in points_by_hex:
            points_by_hex[hex_geom].append(point_idx)
        else:
            points_by_hex[hex_geom] = [point_idx]
    
    # Return only unique geometries with number of points and point indices
    unique_geoms = [geom for geom in hex_with_points['geometry'].unique() if isinstance(geom, Polygon)]
    unique_gdf = gp.GeoDataFrame(geometry=unique_geoms)
    unique_gdf.crs = hex_gdf.crs
    unique_gdf['num_points'] = unique_gdf['geometry'].map(lambda polygon: len(points_by_hex[polygon]))
    unique_gdf['pts_within'] = unique_gdf['geometry'].map(lambda polygon: points_by_hex[polygon])
    
    return unique_gdf
```

```python
sampling_df = sdf11.copy()
sampling_props = np.linspace(0,1,101)
sampled_biomass = []
n_draws = 10

prop_vec = []
mc_biomass = []

for proportion in sampling_props:
    biomass_as_f_prop = []
    for random_seed in range(1,n_draws):
        sample_proportion = proportion
        n_population = len(sampling_df)
        n_samples = int(sample_proportion*n_population)
        biomass_as_f_prop.append(sampling_df.sample(n=n_samples, random_state=random_seed, replace = True)['biomass_MT_co2e_y'].sum() / (n_samples/ n_population))
        prop_vec.append(proportion)
        mc_biomass.append(sampling_df.sample(n=n_samples, random_state=random_seed, replace = True)['biomass_MT_co2e_y'].sum() / (n_samples/ n_population))
        
    sampled_biomass.append(np.nanmean(np.array(biomass_as_f_prop)))
```

```python
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn import metrics
from geopy.distance import great_circle

coordlist = [ (x,y) for x,y in zip(clean_df['lat'],clean_df['lon'])]

kms_per_radian = 6371.0088
epsilon = 0.007 / kms_per_radian
db = DBSCAN(eps=epsilon, min_samples=1, algorithm='ball_tree', metric='haversine').fit(np.radians(coordlist))
cluster_labels = db.labels_
num_clusters = len(set(cluster_labels))
clusters = pd.Series([np.array(coordlist)[cluster_labels == n] for n in range(num_clusters)])
print('Number of clusters: {}'.format(num_clusters))
```

```python
def get_centermost_point(cluster):
    centroid = (MultiPoint(cluster).centroid.x, MultiPoint(cluster).centroid.y)
    centermost_point = min(cluster, key=lambda point: great_circle(point, centroid).m)
    return tuple(centermost_point)

centermost_points = clusters.map(get_centermost_point)
```

```python

from sklearn.datasets import make_classification
from sklearn.cluster import KMeans

n_trees_per_cluster = 20

# define dataset
X = clean_df[['lat','lon']]


# define the model
model = KMeans(n_clusters=int(len(clean_df)/n_trees_per_cluster))
# fit the model
model.fit(X)

# assign a cluster to each example
yhat = model.predict(X)

X['kmeans']=yhat

plt.scatter(X['lat'], X['lon'], c=X['kmeans'],  s=50, alpha=0.8)
plt.colorbar()
```

```python
# from the clusters, make polys based on centroids and bounds 

centers = []
bounds = []

for cluster_id in X['kmeans'].unique():
    # select cluster
    cluster = X[X['kmeans'] == cluster_id]
    # get geom for bounds 
    cgeom = [Point(xy) for xy in zip(cluster['lon'], cluster['lat'])]
    cgdf = gp.GeoDataFrame(cluster, crs="EPSG:4326", geometry=cgeom)
    bounds.append(shapely.geometry.box(*cgdf.total_bounds))
    # get mean for centers 
    meanlat = cluster['lat'].mean()
    meanlon = cluster['lon'].mean()
    geometry = [Point([meanlon,meanlat])]
    tdf = pd.DataFrame([cluster_id])
    tdf.columns = ['cluster_id']
    ctr_gdf = gp.GeoDataFrame(tdf, crs="EPSG:4326", geometry=geometry)
    centers.append(ctr_gdf)
```

```python

ax = pd.concat(centers).reset_index().buffer(0.0002).plot(facecolor = 'none', figsize = (7,7))
pt_gdf.plot( color = 'blue', markersize = 3, ax = ax)
plt.show()
```

```python
plots = pd.concat(centers).reset_index().buffer(0.0002)
print("N pts = {}".format(str(len(pt_gdf))))
print("N polys = {}".format(str(len(plots))))
print("avg pts / poly = {}".format(len(pt_gdf)/len(plots)))
plots = plots.set_crs("EPSG:4326")
ecrs = plots.estimate_utm_crs()  
print("grid cell size = {} meters".format(np.sqrt(plots.to_crs(ecrs).area.iloc[0])))
```

# irregular polygons

```python

def create_irregular_polygons(points_gdf, num_points_per_polygon=10):
    """
    Creates a set of irregular polygons that covers all points in a geopandas geodataframe
    and contains approximately num_points_per_polygon points per polygon.
    
    Args:
    - points_gdf: geopandas geodataframe of points
    - num_points_per_polygon: number of points to be contained in each polygon
    
    Returns:
    - geopandas geodataframe of polygons
    """
    
    # get the coordinates of the points
    coords = np.array(points_gdf.geometry.apply(lambda x: [x.x, x.y]).tolist())
    
    # cluster the points into num_clusters clusters
    num_clusters = int(len(points_gdf) / num_points_per_polygon)
    kmeans = KMeans(n_clusters=num_clusters, random_state=0).fit(coords)
    labels = kmeans.predict(coords)
    
    # create an empty list to store the polygons
    polygons = []
    
    # iterate over each cluster of points
    for i in range(num_clusters):
        # get the coordinates of the points in the cluster
        cluster_coords = coords[labels == i]
        
        # create a convex hull from the points
        cluster_hull = cluster_coords[ConvexHull(cluster_coords).vertices]
        
        # create a polygon from the convex hull
        cluster_polygon = Polygon(cluster_hull)
        
        # add the polygon to the list of polygons
        polygons.append(cluster_polygon)
                
    # create a geopandas geodataframe from the polygons
    polygons_gdf = gp.GeoDataFrame(geometry=polygons)
    
    return polygons_gdf
```

```python
polys = create_irregular_polygons(pt_gdf)
```

```python
ax = polys.buffer(0.00005).plot(facecolor = 'none', figsize = (7,7))
pt_gdf.plot( color = 'blue', markersize = 3, ax = ax)
plt.show()
```

```python
print("N pts = {}".format(str(len(pt_gdf))))
print("N polys = {}".format(str(len(polys))))
print("avg pts / poly = {}".format(len(pt_gdf)/len(polys)))
plots = polys.set_crs("EPSG:4326")
ecrs = plots.estimate_utm_crs()  
print("poly mean area = {} sq meters".format(plots.to_crs(ecrs).area.mean()))
```
