# Uncertainty equations

### Margin of Error (MOE) 
$MOE_{y} = z_{y} * \sqrt( \frac{\sigma}{n} )$

where:
MOE = margin of error <br>
y = confidence interval <br>
$z_{y}$ = quantile <br>
$\sigma$ = standard deviation <br>
n = sample size

### Number of samples (n)

See https://cdm.unfccc.int/methodologies/ARmethodologies/tools/ar-am-tool-03-v2.1.0.pdf

$ n =  \frac { N * t_{val}^2 * {[\sum_{i}w_i * s_i] } ^2 } {N * E^2 + t_{val}^2 * \sum_{i=1}w_i * s_i^2} $  

where: <br>
n = Number of samples required for estimation of biomass stocks within the project boundary; dimensionless <br>
N = N Total number of possible samples within the project boundary (i.e. the sampling space or the population); dimensionless <br>
$t_{val}$ = Two-sided Student√≠s t-value, at infinite degrees of freedom, for the required confidence level; dimensionless <br>
$w_i$ = Relative weight of the area of stratum i (i.e. the area of the stratum i divided by the project area); dimensionless <br>
$s_i$ = Estimated standard deviation of biomass stock in stratum i; t d.m. (or t d.m. ha-1) <br>
E = Acceptable margin of error (i.e. one-half the confidence interval) in estimation of biomass
stock within the project boundary; t d.m. (or t d.m. ha-1), i.e. in the units used for i <br> 
i = 1, 2, 3, ... biomass stock estimation strata within the project boundary

### Equations for sampling and measurement error

$ MAE =  \frac {\sum_{i=1}^{n} |y_i - x_i|} {n} $ 

where $y_i$ are the true values and $x_i$ are the predicted values

$ ME =  \frac {\sum_{i=1}^{n} y_i - x_i} {n} $ 

where $y_i$ are the true values and $x_i$ are the predicted values

$ MAPE = \frac {1}{n} \sum_{i=1}^{n}  \frac {|y_i - x_i|}{y_i} * 100$

where $y_i$ are the true values and $x_i$ are the predicted values

$ MPE = \frac {1}{n} \sum_{i=1}^{n}  \frac {y_i - x_i}{y_i} * 100$

where $y_i$ are the true values and $x_i$ are the predicted values

$R^2 =  1 - \frac {\sum_{i=1}^{n} (x_i - \hat{x})^2}{\sum_{i=1}^{n} (x_i - \bar{x})^2 }$

where $x_i$ are the true values, $\hat{x}$ are the predicted values, and $\bar{x}$ is the mean value

$RMSE =  \sqrt{{ \frac{{\sum_{i=1}^{n} (x_i - \hat{x})^2}} {n} }}$

where $x_i$ and $\hat{x}$ are the same as above, and n is the number of samples

$bias =  \frac {1}{n} {\sum_{i=1}^{n} (x_i - \bar{x})^2 }$

$ubRMSE =  \frac {1}{n} {\sum_{i=1}^{n} [(x_i - \bar{x}) -  (\hat{x_i} - \bar{\hat{x}})]^2 }$

Concordance Correlation Coeff ($\rho_c$) <br> 

$ \rho_c = \frac{2 \rho \sigma_x \sigma_y}{\sigma_{x}^{2} + \sigma_{y}^{2} + (\mu_{x} - \mu_{y})^2}$

IntraClass Correlation <br>

$ r = \frac{1}{Ns^2} \sum_{i=1}^{n} (x_{n,1} - \bar{x}) (x_{n,2} - \bar{x})$ 

where 

$ \bar{x} = \frac{1}{2N} \sum_{i=1}^{n} (x_{n,1} + x_{n,2}) $ 

$ s^2 = \frac{1}{2N} [ \sum_{i=1}^{n} (x_{n,1} - \bar{x})^2 +  \sum_{i=1}^{n} (x_{n,2} - \bar{x})^2 ] $ 


## Statistical Significance Tests

#### Two sided T test
Null hypothesis: The means are equal <br> 
Alternative hypothesis: The means are NOT equal

### Wilcoxon Signed Rank Test
Null hypothesis: Two related paired samples come from the same distribution.<br> 
Alternative hypothesis: Samples come from different distributions.

### Z Test
Null hypothesis: Population means are equal<br> 
Alternative hypothesis: Population means not equal

### Chi Squared Test
Null hypothesis: There are no differences between the classes in the population Alternative <br> 
hypothesis: There are differences between the classes in the population
